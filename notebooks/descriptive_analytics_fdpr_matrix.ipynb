```python
# %% [markdown]
# # ⚖️ Insight 3: Regulatory Exposure - FDPR Matrix
# 
# **Purpose**: Identify financial exposure under FDPR by product type and destination country
# 
# **Key Questions**:
# - Which ECCN codes have the highest FDPR exposure?
# - Which destination countries receive the most FDPR items?
# - What is the total financial exposure by country?

# %% [code]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("="*70)
print("⚖️ INSIGHT 3: REGULATORY EXPOSURE - FDPR MATRIX")
print("="*70)

# %% [code]
# Load the dataset with risk flags
# Try multiple possible paths
possible_paths = [
    '../us_export_compliance_data_with_risk.csv',  # If running from notebooks folder
    'us_export_compliance_data_with_risk.csv',     # If running from main folder
    '../data/us_export_compliance_data_with_risk.csv',  # If in data folder
]

df = None
for path in possible_paths:
    try:
        df = pd.read_csv(path, parse_dates=['shipment_date', 'clearance_date'])
        print(f"✅ File found at: {path}")
        break
    except FileNotFoundError:
        continue

if df is None:
    print("❌ Could not find us_export_compliance_data_with_risk.csv")
    print("Current working directory:", os.getcwd())
    print("Files in current directory:")
    print(os.listdir('.'))
    raise FileNotFoundError("CSV file not found")

print(f"\n📦 Analyzing {len(df):,} shipments with dates from {df['shipment_date'].min().strftime('%Y-%m-%d')} to {df['shipment_date'].max().strftime('%Y-%m-%d')}")

# %% [markdown]
# ## FDPR Exposure Analysis

# %% [code]
# Filter to FDPR jurisdiction items
fdpr_dataset = df[df['fdpr_jurisdiction_flag'] == True]

if len(fdpr_dataset) > 0:
    print(f"\n📊 FDPR Jurisdiction Items: {len(fdpr_dataset):,} shipments")
    
    # Group by ECCN and consignee country
    eccn_exposure = fdpr_dataset.groupby(['eccn', 'consignee_country']).agg({
        'shipment_id': 'count',
        'total_value_usd': 'sum',
        'item_quantity': 'sum'
    }).round(2)
    
    eccn_exposure.columns = ['intercepted_shipments', 'total_exposed_value_usd', 'total_quantity']
    eccn_exposure = eccn_exposure.sort_values('total_exposed_value_usd', ascending=False)
    
    print("\n📊 FDPR EXPOSURE MATRIX (Top 15):")
    display(eccn_exposure.head(15))
    
    # Calculate total exposure
    total_exposure = eccn_exposure['total_exposed_value_usd'].sum()
    total_fdpr_shipments = eccn_exposure['intercepted_shipments'].sum()
    print(f"\n💰 TOTAL FINANCIAL EXPOSURE UNDER FDPR: ${total_exposure:,.2f}")
    print(f"📦 TOTAL SHIPMENTS UNDER FDPR JURISDICTION: {total_fdpr_shipments}")
else:
    print("No FDPR jurisdiction items found in dataset.")

# %% [markdown]
# ## Analysis by Destination Country

# %% [code]
# Group by destination country
country_exposure = fdpr_dataset.groupby('consignee_country').agg({
    'shipment_id': 'count',
    'total_value_usd': 'sum',
    'item_quantity': 'sum'
}).reset_index()

country_exposure.columns = ['Country', 'Shipments', 'Total_Exposure', 'Total_Quantity']
country_exposure = country_exposure.sort_values('Total_Exposure', ascending=False)

print("\n📊 FDPR EXPOSURE BY DESTINATION COUNTRY:")
display(country_exposure.head(10))

# Calculate top country concentration
top_country = country_exposure.iloc[0]['Country']
top_exposure = country_exposure.iloc[0]['Total_Exposure']
top_percentage = (top_exposure / total_exposure * 100)

print(f"\n⚠️ HIGHEST EXPOSURE COUNTRY: {top_country}")
print(f"   Total Exposure: ${top_exposure:,.2f}")
print(f"   Percentage of Total: {top_percentage:.1f}%")

# %% [markdown]
# ## Analysis by ECCN Code

# %% [code]
# Group by ECCN
eccn_exposure_summary = fdpr_dataset.groupby('eccn').agg({
    'shipment_id': 'count',
    'total_value_usd': 'sum',
    'item_quantity': 'sum'
}).reset_index()

eccn_exposure_summary.columns = ['ECCN', 'Shipments', 'Total_Exposure', 'Total_Quantity']
eccn_exposure_summary = eccn_exposure_summary.sort_values('Total_Exposure', ascending=False)

print("\n📊 FDPR EXPOSURE BY ECCN CODE:")
display(eccn_exposure_summary)

# %% [markdown]
# ## Risk Matrix: ECCN vs Country (Heatmap)

# %% [code]
# Create pivot table for heatmap
heatmap_data = pd.pivot_table(
    fdpr_dataset,
    values='total_value_usd',
    index='eccn',
    columns='consignee_country',
    aggfunc='sum',
    fill_value=0
)

print("\n📊 FDPR RISK MATRIX (ECCN × Country):")
print("   This matrix shows exposure dollars at the intersection of ECCN and destination country")

# Show the matrix with formatting
display(heatmap_data)

# %% [markdown]
# ## Top Risk Combinations

# %% [code]
# Get top combinations from the detailed matrix
top_combinations = eccn_exposure.head(10)
print("\n📊 TOP 10 HIGHEST-RISK COMBINATIONS (ECCN × Country):")
for idx, row in top_combinations.iterrows():
    eccn, country = idx
    shipments = row['intercepted_shipments']
    exposure = row['total_exposed_value_usd']
    print(f"   • {eccn} → {country}: {shipments} shipments, ${exposure:,.2f}")

# %% [markdown]
# ## Visualization

# %% [code]
# Create output folder if it doesn't exist
if not os.path.exists('../output'):
    os.makedirs('../output')
    print("✅ Created output folder")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. FDPR Exposure by Country (Top 10)
top_countries = country_exposure.head(10)
colors_country = ['red' if i == 0 else 'orange' if i < 3 else 'lightblue' for i in range(len(top_countries))]
axes[0,0].bar(top_countries['Country'], top_countries['Total_Exposure'], color=colors_country)
axes[0,0].set_title('FDPR Exposure by Destination Country', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Country')
axes[0,0].set_ylabel('Total Exposed Value (USD)')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].grid(True, alpha=0.3)

# 2. FDPR Exposure by ECCN Code
top_eccn = eccn_exposure_summary.head(10)
colors_eccn = ['red' if i == 0 else 'orange' if i < 3 else 'lightblue' for i in range(len(top_eccn))]
axes[0,1].bar(top_eccn['ECCN'], top_eccn['Total_Exposure'], color=colors_eccn)
axes[0,1].set_title('FDPR Exposure by ECCN Code', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('ECCN Code')
axes[0,1].set_ylabel('Total Exposed Value (USD)')
axes[0,1].tick_params(axis='x', rotation=0)
axes[0,1].grid(True, alpha=0.3)

# 3. Heatmap - Risk Matrix
try:
    heatmap_top = heatmap_data.iloc[:10, :8]  # Top 10 ECCN, Top 8 Countries
    if len(heatmap_top) > 0 and len(heatmap_top.columns) > 0:
        sns.heatmap(heatmap_top, ax=axes[1,0], annot=True, fmt='.0f', 
                    cmap='Reds', cbar_kws={'label': 'Exposed Value (USD)'})
        axes[1,0].set_title('FDPR Risk Matrix: ECCN vs Country', fontsize=14, fontweight='bold')
        axes[1,0].set_xlabel('Country')
        axes[1,0].set_ylabel('ECCN Code')
except Exception as e:
    axes[1,0].text(0.5, 0.5, 'Heatmap data not available', ha='center', va='center')
    axes[1,0].set_title('FDPR Risk Matrix: ECCN vs Country', fontsize=14, fontweight='bold')

# 4. FDPR Jurisdiction Pie Chart
fdpr_status = df['fdpr_jurisdiction_flag'].value_counts()
labels = ['Under FDPR', 'Not Under FDPR']
colors_pie = ['coral', 'lightblue']
axes[1,1].pie(fdpr_status.values, labels=labels, autopct='%1.1f%%', colors=colors_pie, startangle=90)
axes[1,1].set_title('FDPR Jurisdiction Status', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../output/insight3_fdpr_exposure.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n💾 Saved: output/insight3_fdpr_exposure.png")

# %% [markdown]
# ## Executive Summary

# %% [code]
print("\n" + "="*70)
print("📊 EXECUTIVE SUMMARY - FDPR EXPOSURE")
print("="*70)

print(f"""
   REGULATORY EXPOSURE:
   ├─ Total FDPR Financial Exposure: ${total_exposure:,.2f}
   ├─ Total FDPR Shipments: {total_fdpr_shipments:,}
   └─ Total FDPR Items: {fdpr_dataset['item_quantity'].sum():,}

   TOP EXPOSURE BY COUNTRY:
   ├─ 1. {country_exposure.iloc[0]['Country']}: ${country_exposure.iloc[0]['Total_Exposure']:,.2f}
   ├─ 2. {country_exposure.iloc[1]['Country']}: ${country_exposure.iloc[1]['Total_Exposure']:,.2f}
   └─ 3. {country_exposure.iloc[2]['Country']}: ${country_exposure.iloc[2]['Total_Exposure']:,.2f}

   TOP EXPOSURE BY ECCN:
   ├─ 1. {eccn_exposure_summary.iloc[0]['ECCN']}: ${eccn_exposure_summary.iloc[0]['Total_Exposure']:,.2f}
   ├─ 2. {eccn_exposure_summary.iloc[1]['ECCN']}: ${eccn_exposure_summary.iloc[1]['Total_Exposure']:,.2f}
   └─ 3. {eccn_exposure_summary.iloc[2]['ECCN']}: ${eccn_exposure_summary.iloc[2]['Total_Exposure']:,.2f}

   RECOMMENDATIONS:
   1. Prioritize manual review of FDPR shipments to {top_country} (${top_exposure:,.2f})
   2. Conduct compliance audit for {eccn_exposure_summary.iloc[0]['ECCN']} classified items
   3. Implement automated FDPR detection for high-risk ECCN-country combinations
""")

print("\n✅ INSIGHT 3 COMPLETE!")
print("="*70)
```
